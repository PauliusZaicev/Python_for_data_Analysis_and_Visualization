```python
import numpy as np

#for a visualisation
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#grid
points = np.arange(-5,5,0.01)
```

```python
#create matrix, from -5 to 5 by 0.1
dx, dy = np.meshgrid(points,points)
#in liens
dx
```

```python
#in columns
dy
```

```python
#evaluating function

z = (np.sin(dx) + np.sin(dy))
z
```

```python
#plot image, sciamatrix
plt.imshow(z)
```

```python
#display color meaning
plt.imshow(z)
plt.colorbar()
plt.title('Plot for sin(x)+ sin(y)')
```

```python
#nupy where

A = np.array ([1,2,3,4])

B = np.array ([100,200,300,400])
```

```python
#create bulian array

condition = np.array([True, True, False, False])
```

```python
#list comprehension

#use ZIP to input values

answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
```

```python
answer
```

```python
#short form of a previuos example
answer2 = np.where(condition, A,B)
```

```python
answer2
```
